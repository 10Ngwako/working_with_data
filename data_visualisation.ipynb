**VISUALISING THE DATA**

*This shows the distribution of age within the dataset*

```python
# Histogram

plt.figure(figsize=(5, 4)) # size of the plot
sns.histplot(df['age'], bins=30, kde=True, color='green', label = 'age')
plt.title('Distribution of Age') #title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

*Scatter plot below displays the relationship between age and annual salary.*

```python

# Scatter Plot

plt.figure(figsize=(5, 4)) # size of the plot
sns.scatterplot(x='age', y='annual Salary', data=df, color='skyblue', label = 'Data points')
plt.title('CORRELATION BETWEEN AGE AND ANNUAL SALARY')
plt.xlabel('AGE')
plt.ylabel('ANNUAL SALARY')
plt.legend()
plt.show()
```

*The line chart below shows the trend and changes of the dataset over time.*

```python

# Line Chart

plt.figure(figsize=(13, 7)) #size
sns.lineplot(x='age', y='net worth', data=df, marker='s', color='blue', )
plt.title('NET WORTH OF INDIVIDUALS BASED ON THEIR DISTINCT AGES')
plt.xlabel('AGE')
plt.ylabel('NET WORTH')
plt.show()
```

*The heatmap below analyzed the relationship between the distinct columns given.*

```python

# Heatmap

# Selecting the columns for heatmap
columns = ['country', 'gender', 'age']

# Label encoding for categorical variables
for col in columns:
    df[col] = df[col].astype('category').cat.codes

# Plotting the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df[columns].corr(), annot=True, cmap='coolwarm')
plt.title('CORRELATION HEATMAP FOR CAR PURCHASING')
plt.show()
```
